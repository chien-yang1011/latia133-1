```python
def a(x,y):
  result = x+y*2
  return result

b=a(1,2)
c=a(2,3)


print(b)
print(c)
```

```python
def a(x):
  result=0
  while result<10:
    result=result+x

  if result==5:
      return result
  return result

b=a(1)
c=a(2)


print(b)
print(c)
```

```python
def tf(value):
   return 2*value-1
test_values=[0.0,0.0011,0.25,0.2,0.54,0.75,1.0]
scaled_values=[tf(value) for value in test_values]

print("original values:",test_values)
print("scaled values:",scaled_values)
```

```python
def a(n):
  if n ==0 or n==1:
    return 1
  else:
    return n + a(n-1)
print(a(3))

```
